<h2>Importamos librerías</h2>

```python
import numpy as np
import random
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
%matplotlib inline
```

<h2>Generador de distribución para regresión lineal simple</h2>
<p>Dentro de este apartado nosotros vamos a generar el Dataset sobre el cual vamos a trabajar con nuestro modelo para poder obtener los datos</p>
<p>De igual forma se genera  la gráfica sobre la cual vamos a mostrar nuestros datos ya generados</p>

```python
def generador(pendiente, muestras, desviacion):
    
    #Aquí se generan N muestras de distribución de datos para la regresión lineal simple
    X = np.random.random(muestras) * 100
    
    #Generador de ruido gaussiano con desviación típica
    e = np.random.rand(muestras) * desviacion

    #Se obtiene el valor real de Y como X * pendiente + error
    Y = X * pendiente + e

    return X.reshape((muestras,1)), Y.reshape((muestras,1))

#Parámetros de la distribución
desviacion = 200
w = 10
n = 100

X,Y = generador(w, n, desviacion)

plt.scatter(X, Y, alpha = 0.6)
plt.show()
```

<h2>Procesamiento de los datos</h2>
<p>Una vez generados los datos sobre los cuales escribiremos nuestro modelo, toca procesarlos con la librería de SKlearn</p>

```python
#Creamos el modelo de regresión lineal
modelo = linear_model.LinearRegression()

#Entrenamos el modelo con los datos de X y Y
modelo.fit(X,Y)

#Ahora obtenemos el coeficiente w 
print("El valor de b es " + str(modelo.intercept_))
print("El valor de w1 es: " + str(modelo.coef_[0]))

#podemos predecir ocupando el modelo
Y_pred = modelo.predict(X)

# Por último, calculamos el error cuadrático medio y el estadístico R^2
print ("Error cuadrático medio: " + str(mean_squared_error(Y, Y_pred)))
print ("Estadístico R_2: " + str(r2_score(Y, Y_pred)))
```

<h2>Resultado</h2>

<h3>Graficación de la linea resultante</h3>
<p>Ahora lo que debemos de hacer es gráficar la linea resultante y ver de que manera esta se relaciona con los datos iniciales del modelo.</p>

```python
plt.scatter(X, Y, alpha = 0.6)
plt.plot(X, Y_pred, color="red", linewidth=2)
plt.show()
```

<h3>Predicción de valores</h3>
<p>Ahora generaremos predicciones tomando como referencia un valor de entrada para conocer un valor de salida que no se encuentra dentro de los datos reales.</p>

```python
def prediccion(entrada):
    salida = modelo.coef_[0]  * entrada + modelo.intercept_
    return salida[0]

print("Según los datos con un valor de 46 nosotros obtendremos un valor aproximado a " +     str(prediccion(46)) + " según los valores reales")
```
